```python
def gradient_descent():
    x = df['Hours']
    y = df['Scores']
    m = b = 0
    iterations = 1000
    n = len(x)
    learning_rate = 0.001
    
    for i in range(iterations):
        for j in range(0, n):
            y_predicted = m * x[j] + b
            md = (1/float(n))*(x[j]*(y_predicted- y[j]))
            bd = (1/float(n))* (y_predicted -y[j])
            m = m - learning_rate * md
            b = b - learning_rate * bd
    
        print m, b
gradient_descent()
```

```python
theta1, theta0 = gradient_descent()
theta1, theta0
```
